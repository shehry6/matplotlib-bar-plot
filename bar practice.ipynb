# matplotlib bar plot

```python
x = ["python","java","hacking","web","cyber"]
y = [8,2,3,5,6]
z = [60,10,30,40,44]

c = ["r","g","b","y","b"]

plt.xlabel("Course",fontsize = 15)
plt.ylabel("Nos",fontsize = 15)
plt.title("SHEHROZ KHAN",fontsize = 15)

plt.bar(x,y,color =c ,width = 0.5,label="Popular",)
plt.bar(x,z,color ="r",width = 0.5,label="less Popular",)
plt.legend()
plt.show()
```
